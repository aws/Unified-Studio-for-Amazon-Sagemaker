# Parameter Test Notebook
This notebook tests parameter passing from workflow to notebook

```python
# Parameters cell - tagged for papermill injection
mlflow_arn = None
```

```python
print("="*50)
print("PARAMETER TEST RESULTS")
print("="*50)
print(f"Received parameter: {mlflow_arn}")
print(f"Parameter type: {type(mlflow_arn)}")
print(f"Parameter value: '{mlflow_arn}'")
print("="*50)

# Check parameter was passed (not None) and is a valid ARN format
if mlflow_arn is None:
    raise ValueError("❌ FAILED: mlflow_arn parameter was not passed")
elif not isinstance(mlflow_arn, str):
    raise ValueError(f"❌ FAILED: mlflow_arn should be string, got {type(mlflow_arn)}")
elif not mlflow_arn.startswith("arn:aws:sagemaker:"):
    raise ValueError(f"❌ FAILED: mlflow_arn should be a valid SageMaker ARN, got '{mlflow_arn}'")
elif "mlflow-tracking-server" not in mlflow_arn:
    raise ValueError(f"❌ FAILED: mlflow_arn should contain 'mlflow-tracking-server', got '{mlflow_arn}'")
else:
    print("✅ SUCCESS: Parameter passed correctly and is a valid MLflow ARN!")
```
